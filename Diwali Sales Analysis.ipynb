<h3>countplot:</h3>
Displays the count of occurrences of categories in a dataset.
Automatically computes the counts for each category in a categorical variable.
    
<h3>barplot:</h3>
Displays a bar plot for the mean (or another aggregate function) of a numeric variable grouped by a categorical variable.
Requires both a numeric and a categorical variable.


```python
# sns.barplot(x='Age Group',y='Amount',data=AgeGroupWisePurchase)
ax=sns.countplot(data=df,x='Age Group',hue='Gender')  #LEGEND ALSO CALLED IN SEABORN IS "hue"
for bars in ax.containers:
    ax.bar_label(bars)
```

<b><h3>State</h3></b>

```python
StateWiseOrders=df.groupby('State',as_index=False)[['Orders']].sum().sort_values(by='Orders',ascending=False)
StateWiseOrders
```

```python
sns.set(rc={'figure.figsize':(10,5)})
state=sns.barplot(x='State',y='Orders',data=StateWiseOrders)
plt.xticks(rotation=90)  
plt.show()
```

```python
statewiseamount=df.groupby('State',as_index=False)[['Amount']].sum().sort_values(by='Amount',ascending=False)
statewiseamount
```

<b>sometimes state have a more purchasing power more or sometimes diff states more products purchase</b>

```python
sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(x='State',y='Amount',data=statewiseamount)
plt.xticks(rotation=90)
plt.show()
```

<b>From above graphs we can see that most of the orders and total sales/amount are from uttar pradesh,maharashtra and karnataka respectively</b>

<h2>Product Category</h2>

```python
unique_items = df['Product_Category'].unique()
print(unique_items)
```

```python
ax=sns.countplot(data=df,x='Product_Category')
for bar in ax.containers:
    ax.bar_label(bar)
plt.xticks(rotation=90)
plt.show()
```

```python
ProductCategoryAmount=df.groupby('Product_Category',as_index=False)[['Amount']].sum().sort_values(by='Amount',ascending=False)
ProductCategoryAmount
```

<b>Insights:From above graphs cans ee that most of the sold products are from Food,Clothing & Apparel and Footwear & Shoes category</b>
